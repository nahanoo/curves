### Fitting for Comamonas testosteroni using 7.5 mM Potassiumacetate (15 mM of carbon)

```python
# Load data
meta,raw = get_dfs('./')
# Filter data
mask = (meta["species"] == 'Comamonas testosteroni') & (meta["cs_conc"] == 15)
columns = list(set(meta[mask]["linegroup"]))
ct15 = raw[["time"] + columns].dropna()
```

```python
# Plotting raw data
plot_OD(ct15)
```

```python
# Masking data to cut off noise at the end
ct15 = mask_df(ct15, 0, 30)
plot_OD(ct15)
```

```python
# Fitting maximum growth rate using linear regression
v = fit_max_growth_rate(ct15, 0, 5, plot=True)
print(v)
```

```python
# Fitting Km
n0 = get_n0(ct15)
# 15 is carbon concentration
q = get_yield(ct15, 15)
Km = fit_Km(ct15, v, 15)
plot_fit(ct15, Km, v, n0, q, 15)
print(Km)
```
